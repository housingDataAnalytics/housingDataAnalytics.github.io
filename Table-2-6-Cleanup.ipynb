```python
import pandas as pd
import matplotlib.pyplot as plt 
```

# Cleaning Table 2

```python
# Reading in table 2.
LIHTC_table2 = pd.read_excel("2018-LIHTC-Tenant-Tables.xlsx", sheet_name="Table 2")

LIHTC_table2.iloc[0, 1:5] = "Properties"
LIHTC_table2.iloc[0, 5:] = "Tenants"

# Multicolumns indices for the table. 
col_names1 = LIHTC_table2.iloc[0, 1:].values
col_names2 = LIHTC_table2.iloc[1, 1:].values

states = LIHTC_table2.iloc[2:57, 0].values

properties_df = LIHTC_table2.iloc[2:57, 1:].copy()

properties_df.index = states
properties_df.columns = [col_names1, col_names2]

properties_df
```

# Cleaning Table 6

```python
# Reading in table 2.
disabilities_df = pd.read_excel("2018-LIHTC-Tenant-Tables.xlsx", sheet_name="Table 6")

## Parsing data from excel file

# Retrieving row names from first row and removing whitespace.
col_names = disabilities_df.iloc[0, 1:].str.replace(r'\s{2,}', '', regex=True).values
# Retrieving states from first column to use as indexes.
states = disabilities_df.iloc[2:57, 0].values
# Removing extra rows and columns.
disabilities_df = disabilities_df.iloc[2:57, 1:]

disabilities_df.columns = col_names
disabilities_df.index = states

disabilities_df
```

```python
ax = disabilities_df.plot.bar(y=disabilities_df.columns[3])

plt.rcParams["figure.figsize"] = (80,40)
ax.tick_params(axis='both', which='major', labelsize=40)

ax.legend(prop={'size': 65})
plt.show()
```
